```python
"""Лабораторная работа №3: Алгоритмы поиска и сортировки.
Выполнила: Самусева В.А. ЦИБ-251
Задание №2 Блок 1."""

def linear_search_str(arr, target):
    # Перебираем все элементы списка с их индексами
    for i, item in enumerate(arr):
        if item == target:  # Сравнение строк (регистрозависимое)
            return i
    return -1


"""""""""""""""""""Пример использования"""""""""""""""""""
print(linear_search_str(["apple", "banana", "cherry"], "banana"))  # 1
print(linear_search_str(["apple", "banana", "cherry"], "Banana"))  # -1 (разный регистр)
```
