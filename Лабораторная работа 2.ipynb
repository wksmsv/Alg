```python
"""
Лабораторная работа №2: Графы и деревья (Вариант 22)
Один файл с реализацией всех алгоритмов и демонстрацией.
Сложность указана в комментариях к строкам (время выполнения).
"""

# ============================================================
# 1. Работа с графами
# ============================================================

def build_adjacency_matrix(num_vertices, edges):
    """Матрица смежности. Общая сложность: O(V^2 + E), память O(V^2)."""
    # Создание матрицы VxV, заполненной нулями – O(V^2) времени
    matrix = [[0] * num_vertices for _ in range(num_vertices)]  # O(V^2)
    for u, v in edges:                          # O(E) итераций
        matrix[u - 1][v - 1] = 1                # O(1)
        matrix[v - 1][u - 1] = 1                # O(1)
    return matrix                               # O(1)

def build_incidence_matrix(num_vertices, edges):
    """Матрица инцидентности. Общая сложность: O(V*E + E), память O(V*E)."""
    num_edges = len(edges)                      # O(1)
    # Матрица V строк, E столбцов – O(V*E) времени
    matrix = [[0] * num_edges for _ in range(num_vertices)]  # O(V*E)
    for e_idx, (u, v) in enumerate(edges):      # O(E) итераций
        matrix[u - 1][e_idx] = 1                # O(1)
        matrix[v - 1][e_idx] = 1                # O(1)
    return matrix                               # O(1)

def find_connected_components(num_vertices, edges):
    """Компоненты связности через DFS. Общая сложность: O(V + E)."""
    # Построение матрицы смежности – O(V^2 + E)
    adj = build_adjacency_matrix(num_vertices, edges)  # O(V^2 + E)
    visited = [False] * num_vertices            # O(V)
    components = []                             # O(1)

    def dfs(v, comp):
        visited[v] = True                       # O(1)
        comp.append(v + 1)                      # O(1) (амортизированно)
        for u in range(num_vertices):           # O(V) итераций для каждой вершины
            if adj[v][u] == 1 and not visited[u]:  # O(1)
                dfs(u, comp)                    # рекурсивный вызов – в сумме O(V+E)

    for i in range(num_vertices):               # O(V) итераций
        if not visited[i]:                      # O(1)
            comp = []                           # O(1)
            dfs(i, comp)                        # суммарно по всем вызовам O(V+E)
            components.append(comp)             # O(1) (амортизированно)
    return components                           # O(1)

# ============================================================
# 2. Бинарное дерево поиска (BST)
# ============================================================

class BSTNode:
    def __init__(self, key):
        self.key = key                          # O(1)
        self.left = None                        # O(1)
        self.right = None                       # O(1)

class BinarySearchTree:
    def __init__(self):
        self.root = None                        # O(1)

    def insert(self, key):
        """Вставка: в среднем O(log n), в худшем O(n)."""
        if self.root is None:                   # O(1)
            self.root = BSTNode(key)            # O(1)
        else:
            self._insert_rec(self.root, key)    # O(log n) средн., O(n) худш.

    def _insert_rec(self, node, key):
        if key < node.key:                      # O(1)
            if node.left is None:               # O(1)
                node.left = BSTNode(key)        # O(1)
            else:
                self._insert_rec(node.left, key)  # рекурсивно: O(log n) ср.
        elif key > node.key:                    # O(1)
            if node.right is None:              # O(1)
                node.right = BSTNode(key)       # O(1)
            else:
                self._insert_rec(node.right, key) # рекурсивно: O(log n) ср.

    def search(self, key):
        """Поиск: в среднем O(log n), в худшем O(n)."""
        return self._search_rec(self.root, key) # O(log n) ср.

    def _search_rec(self, node, key):
        if node is None:                        # O(1)
            return False                        # O(1)
        if key == node.key:                     # O(1)
            return True                         # O(1)
        elif key < node.key:                    # O(1)
            return self._search_rec(node.left, key)  # рекурсивно
        else:
            return self._search_rec(node.right, key) # рекурсивно

    def delete(self, key):
        """Удаление: в среднем O(log n), в худшем O(n)."""
        self.root = self._delete_rec(self.root, key)  # O(log n) ср.

    def _delete_rec(self, node, key):
        if node is None:                        # O(1)
            return None                         # O(1)
        if key < node.key:                      # O(1)
            node.left = self._delete_rec(node.left, key)  # рекурсивно
        elif key > node.key:                    # O(1)
            node.right = self._delete_rec(node.right, key) # рекурсивно
        else:
            # узел найден
            if node.left is None:               # O(1)
                return node.right               # O(1)
            elif node.right is None:            # O(1)
                return node.left                # O(1)
            else:
                # два потомка: поиск преемника – O(log n) в среднем
                succ = self._min_value_node(node.right)  # O(log n)
                node.key = succ.key             # O(1)
                node.right = self._delete_rec(node.right, succ.key)  # рекурсивно
        return node                             # O(1)

    def _min_value_node(self, node):
        current = node                          # O(1)
        while current.left is not None:         # максимум O(log n) итераций (высота)
            current = current.left              # O(1)
        return current                          # O(1)

    def inorder(self):
        """Симметричный обход: O(n)."""
        result = []                             # O(1)
        self._inorder_rec(self.root, result)    # O(n)
        return result                           # O(1)

    def _inorder_rec(self, node, result):
        if node:                                # O(1)
            self._inorder_rec(node.left, result)  # рекурсивно
            result.append(node.key)             # O(1) амортизированно
            self._inorder_rec(node.right, result) # рекурсивно

# ============================================================
# 3. Пирамидальная сортировка (Heap Sort)
# ============================================================

def heapify(arr, n, i):
    """Восстановление max-heap. Сложность O(log n)."""
    largest = i                                 # O(1)
    left = 2 * i + 1                            # O(1)
    right = 2 * i + 2                           # O(1)

    if left < n and arr[left] > arr[largest]:   # O(1)
        largest = left                          # O(1)
    if right < n and arr[right] > arr[largest]: # O(1)
        largest = right                         # O(1)

    if largest != i:                            # O(1)
        arr[i], arr[largest] = arr[largest], arr[i]  # O(1)
        heapify(arr, n, largest)                # рекурсивный вызов, глубина O(log n)

def heap_sort(arr):
    """Пирамидальная сортировка. Общая сложность O(n log n)."""
    a = arr.copy()                              # O(n)
    n = len(a)                                  # O(1)

    # Построение кучи – O(n log n) (точнее O(n), но для наглядности)
    for i in range(n // 2 - 1, -1, -1):         # O(n) итераций
        heapify(a, n, i)                        # O(log n) каждый → O(n log n) в целом

    # Извлечение элементов – O(n log n)
    for i in range(n - 1, 0, -1):               # O(n) итераций
        a[0], a[i] = a[i], a[0]                 # O(1)
        heapify(a, i, 0)                        # O(log i) ~ O(log n)

    return a                                    # O(1)

# ============================================================
# 4. Демонстрация для варианта 22
# ============================================================

def print_matrix(title, mat):
    print(title)                                # O(1)
    for row in mat:                             # O(V) для матрицы VxV
        print(" ".join(f"{x:2}" for x in row))  # O(V) на строку → O(V^2) всего
    print()                                     # O(1)

def main():
    # ------------------- Граф -------------------
    num_v = 5                                   # O(1)
    edges = [(1, 2), (2, 3), (3, 4), (4, 5)]    # O(1)

    adj = build_adjacency_matrix(num_v, edges)  # O(V^2 + E)
    inc = build_incidence_matrix(num_v, edges)  # O(V*E + E)
    comps = find_connected_components(num_v, edges)  # O(V + E)

    print("\n>>> ГРАФ: 5 вершин, рёбра:", edges)
    print_matrix("Матрица смежности:", adj)
    print_matrix("Матрица инцидентности:", inc)
    print("Компоненты связности:", comps)

    # ------------------- BST -------------------
    bst = BinarySearchTree()                    # O(1)
    elements = [29, 20, 36, 18, 25, 32, 40]     # O(1)
    for el in elements:                         # O(n) для n=7
        bst.insert(el)                          # O(log n) в среднем → O(n log n)

    print("Исходные элементы:", elements)
    print("Обход (inorder):", bst.inorder())    # O(n)
    print(f"Поиск 36: {'найден' if bst.search(36) else 'не найден'}")  # O(log n)
    bst.delete(29)                              # O(log n) сред.
    print("После удаления 29:", bst.inorder())  # O(n)

    # ------------------- Heap Sort -------------------
    arr = [29, 20, 36, 18, 25, 32, 40]          # O(1)
    print("Исходный массив:", arr)
    sorted_arr = heap_sort(arr)                 # O(n log n)
    print("Отсортированный массив:", sorted_arr)

    print("\n" + "=" * 60)
    print("Работа программы завершена.")

if __name__ == "__main__":
    main()
```
