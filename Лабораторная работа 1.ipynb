```python
# Лабораторная работа №1

"""Автор: Самусева Виктория, группа ЦИБ-251
Вариант: 22 – Дек. Реализация скользящего максимума в массиве (Sliding Window Maximum)

## Описание задания
Дан массив целых чисел и размер окна `k`. Необходимо найти максимум в каждом непрерывном подмассиве длины `k`.
Например:
Вход: `[1, 3, -1, -3, 5, 3, 6, 7]`, `k = 3`
Выход: `[3, 3, 5, 5, 6, 7]`"""

from typing import List, Any, Optional

# ---------- 1. КЛАСС Deque ----------
class Deque:
    """Двусторонняя очередь на списке."""

    def __init__(self):
        self._items: List[Any] = []          # O(1)

    def is_empty(self) -> bool:
        return len(self._items) == 0         # O(1)

    def size(self) -> int:
        return len(self._items)              # O(1)

    def append_left(self, item: Any) -> None:
        self._items.insert(0, item)          # O(n) – сдвиг элементов

    def append_right(self, item: Any) -> None:
        self._items.append(item)             # O(1) амортизированно

    def pop_left(self) -> Optional[Any]:
        if self.is_empty():                  # O(1)
            return None
        return self._items.pop(0)            # O(n) – сдвиг

    def pop_right(self) -> Optional[Any]:
        if self.is_empty():                  # O(1)
            return None
        return self._items.pop()             # O(1)

    def peek_left(self) -> Optional[Any]:
        if self.is_empty():                  # O(1)
            return None
        return self._items[0]                # O(1)

    def peek_right(self) -> Optional[Any]:
        if self.is_empty():                  # O(1)
            return None
        return self._items[-1]               # O(1)

# ---------- 2. АЛГОРИТМ СКОЛЬЗЯЩЕГО МАКСИМУМА ----------
def sliding_window_maximum(arr: List[int], k: int) -> List[int]:
    # Проверки граничных условий
    if not arr or k <= 0:                    # O(1)
        return []                            # O(1)

    n = len(arr)                             # O(1)
    if k >= n:                               # O(1)
        return [max(arr)]                    # O(n) – только в этом особом случае

    dq = Deque()                             # O(1)
    result = []                              # O(1)

    for i in range(n):                       # Всего O(n) итераций
        # 1) Удалить индексы, вышедшие за окно
        while (not dq.is_empty()             # O(1)
               and dq.peek_left() <= i - k): # O(1) на peek_left
            dq.pop_left()                    # O(n) в данной реализации, но суммарно O(n) за всю работу
                                             # каждый индекс удаляется не более 1 раза

        # 2) Удалить индексы с меньшими/равными значениями
        while (not dq.is_empty()             # O(1)
               and arr[dq.peek_right()] <= arr[i]): # O(1) на доступ по индексу
            dq.pop_right()                   # O(1) операция, суммарно O(n)

        dq.append_right(i)                   # O(1)

        # 3) Начиная с заполненного окна, сохраняем максимум
        if i >= k - 1:                       # O(1)
            result.append(arr[dq.peek_left()]) # peek_left O(1), append O(1)

    return result                            # O(1)

    # ---------- 3. ПРИМЕР ЗАПУСКА ----------
if __name__ == "__main__":
    arr = [1, 3, -1, -3, 5, 3, 6, 7]
    k = 3
    print(sliding_window_maximum(arr, k))    # [3, 3, 5, 5, 6, 7]
```
